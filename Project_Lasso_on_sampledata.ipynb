```python
# the numerical columns
oth_cols = data.loc[:, data.dtypes != np.object]
oth_cols.columns
```

```python
other_cols = ['Attribute 3', 'Attribute 4', 'Metric1',
       'Attribute 6', 'Attribute 6_Type 1 %', 'Attribute 6_Type 2 %',
       'Attribute 6_Type 3 %', 'Attribute 6_Type 4 %', 'Metric 2 %', 'Metric2']
```

```python
# Label categorical columns with LabelEncoder

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_X = LabelEncoder()

X_labelled = X[cat_cols].apply(labelencoder_X.fit_transform)
X_labelled
```

```python
# df.apply(LabelEncoder().fit_transform)
```

```python
# then convert all the labelled columns into 0s and 1s
onehotencoder_X = OneHotEncoder()

X_coded = onehotencoder_X.fit_transform(X_labelled).toarray()
X_coded
```

```python
X_done = pd.DataFrame(X_coded)
X_done
```

```python
dataframe_right = data[other_cols]
dataframe_right.reset_index(inplace=True)
dataframe_right
```

```python
# combined the processed columns with the rest of the columns to form a new dataframe
data_precessed = pd.concat([X_done, dataframe_right], axis=1)
data_precessed
```

```python
# save the processed dataframe in a new .csv file
data_precessed.to_csv('sampledata_processed.csv')
```

```python
# all variable vaules
Xp = data_precessed.iloc[:, 1:-1]
```

```python
# all target vaules
yp = data_precessed.iloc[:, -1]
```

```python
# train test split by sklearn
import sklearn.model_selection
(Xtrain, Xtest, ytrain, ytest) = sklearn.model_selection.train_test_split(Xp, yp)
```

```python
# Scale all data
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
Xtrain = sc_X.fit_transform(Xtrain)
Xtest = sc_X.transform(Xtest)
```

```python
# search the best Alpha for Lasso model
lasso_cv = sklearn.linear_model.LassoCV(alphas=[0.0005, 
                                                0.0007,
                                                0.0009,
                                                0.001,
                                                0.0011,
                                                0.0025,
                                                0.01, 0.1, 1.0, 1.1, 2.0, 3.0, 10.0 , 11,20,30, 35, 37, 37.5, 39
                                                , 40, 50, 100],
                                       normalize=True,
                                       cv=78,
                                        n_jobs=-1
                                       )
lasso_cv.fit(Xp, yp)
print ("Best alpha is ", lasso_cv.alpha_)
print ("Best intercept is", lasso_cv.intercept_)
```

```python
pd.DataFrame(lasso_cv.coef_, Xp.columns, columns = ['lasso_score']).lasso_score.sort_values(ascending=False)
```

```python
# Use the best alpha 35 in Lasso model and get the lasso score
lasso2 = sklearn.linear_model.Lasso(alpha=35)
lasso2.fit(Xtrain, ytrain)
print (lasso2.score(Xtest, ytest))
```

```python
lasso2.coef_
```

```python
# get the most relavent columns
abs(pd.DataFrame(lasso2.coef_, Xp.columns, columns = ['lasso_score'])).lasso_score.sort_values(ascending=False)
```

```python
lasso2.intercept_
```

```python
# compare to see what the score of a dummy model, and the dummy model performs really bad

import sklearn.dummy
dummy = sklearn.dummy.DummyRegressor()
dummy.fit(Xtrain, ytrain)
dummy.score(Xtest, ytest)
```
